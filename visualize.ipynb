```python
import pandas as pd
```

```python
df=pd.read_csv("population.csv")
df
```

```python
df["Year"] = df["Year"].astype(int) 
```

```python
df.set_index("Year",inplace=True)
```

```python
df.sort_index(inplace=True)
```

```python
df
```

```python
df["Population"] = df["Population"].str.replace(",", "").astype(float)
df["Urban Population"] = df["Urban Population"].str.replace(",", "").astype(float)
df["World Population"] = df["World Population"].str.replace(",", "").astype(float)
df["Migrants (net)"] = df["Migrants (net)"].str.replace(",", "").astype(float)
```

```python
df["Population"].plot(kind="bar",title='Population Growth Over Time', xlabel='Year', ylabel='Population (in billions)')
```

```python
df[["Population","World Population"]].plot(kind="bar" ,stacked=True,title="World vs India Population Over Years")
```

```python
df[["Urban Population","Population"]].plot(kind="bar",stacked=True,title='Urban vs Rural Population Over Years')
```

```python
df.loc[2024,["Population","World Population"]].plot(kind="pie",title="Percentage of Population in 2024",autopct='%1.1f%%', startangle=90)
```

```python
df.loc[2024,["Population","Urban Population"]].plot(kind="pie",title="Percentage of Urban Population in 2024",autopct='%1.1f%%', startangle=90)
```

```python
df["Fertility Rate"].plot(title='Fertility Rate Trends Over Time', xlabel='Year', ylabel='Fertility Rate')
```

```python
df['Migrants (net)'].plot(kind='bar', title='Net Migration Trends Over Time', ylabel='Net Migration')
```
